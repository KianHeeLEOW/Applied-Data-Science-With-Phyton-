```python
import glob
import pandas as pd
import os
import csv
import numpy as np
```

```python
singaporedf=pd.read_csv('Singapore Data Quality.csv')
```

```python
singaporedf_clean=singaporedf[(singaporedf['year']>=2011)]
singaporedf_clean=singaporedf_clean[['year','pm2.5_mean','carbon_monoxide_2nd_maximum_8hourly_mean','nitrogen_dioxide_mean']]
singaporedf_clean.columns = ['Year', 'PM2.5','Carbon monoxide','Nitrogen dioxide']
singaporedf_clean['State Name'] = 'Singapore'
singaporedf_clean
```

```python
usdf=pd.read_csv('US Data Quality.csv')
usdf
```

```python
usdf_clean=usdf[(usdf['Parameter Name']=='PM2.5 - Local Conditions')|(usdf['Parameter Name']=='Nitrogen dioxide (NO2)')| (usdf['Parameter Name']=='Carbon monoxide')]
usdf_clean=usdf_clean[(usdf_clean['Year']<=2014)]
```

```python
usdf_clean
```

```python
usdf_tidy=usdf_clean.groupby(['State Name', 'Year','Parameter Name'], as_index=False).mean()
```

```python
singaporedf_clean2 = pd.melt(singaporedf_clean, id_vars=["State Name","Year"], 
                  var_name="Parameter Name", value_name="Arithmetic Mean")
```

```python
singaporedf_clean2
```

```python
usdf_tidy
```

```python
#Append the 2 data frames together
combineddf=singaporedf_clean2.append(usdf_tidy,ignore_index=True)
```

```python
#Standardise the Paramter Name fields
di = {"PM2.5 - Local Conditions": "PM2.5", "Nitrogen dioxide (NO2)": "Nitrogen dioxide"}
combineddf=combineddf.replace({"Parameter Name": di})
```

```python
combineddf
```

```python
#filter other states except for Singapore, New York, Alaska, Texas and Washington
#View mean PM2.5 index
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(10, 10))
filtereddf=combineddf[(combineddf['State Name']=='Singapore')| (combineddf['State Name']=='New York')| (combineddf['State Name']=='Texas')| (combineddf['State Name']=='Washington')| (combineddf['State Name']=='Alaska')]
sns.pointplot(x='Year', y='Arithmetic Mean', hue='State Name', data=filtereddf[(combineddf['Parameter Name']=='PM2.5')])
sns.despine(fig)
plt.title('PM2.5 Index')
plt.legend();
plt.show()
```
